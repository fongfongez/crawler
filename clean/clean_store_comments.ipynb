```python
import  re
# 清理身分欄位
def prase_user_info(text):
    text = str(text) if not isinstance(text, str) else text
    # 檢查是不是字串，不是的話就是字串類型，才能判斷
    is_local_guide = '在地嚮導' in text
    review_cnt = 0
    photo_cnt = 0
    
    reviews_match = re.search(r"([\d,]+)\s*則評論", text) # 找尋(有多位數字含,為第一組)空格則評論 -> str
    if reviews_match:
        review_cnt = int(reviews_match.group(1).replace(",",""))  # 第一組數字,取代成空字串轉成整數
    else:
        review_cnt == 0

    photos_match = re.search(r"([\d,]+)\s*張相片", text) # 找尋(有多位數字含,為第一組)空格張相片 -> str
    if photos_match:
        photo_cnt = int(photos_match.group(1).replace(",","")) # 第一組數字,取代成空字串轉成整數
    else:
        photo_cnt == 0
    
    return pd.Series({
        "is_local_guide": is_local_guide,
        "review_cnt": review_cnt,
        "photo_cnt": photo_cnt
    })
```

```python
df['留言者身分'].apply(prase_user_info)
```

```python
df[['local_guide', 'review_count', 'photo_count']] = df['留言者身分'].apply(prase_user_info)
df
```

```python
def split_time(text):
    time_match = re.split(r"(\d+)(.+?)前", text)
    time_num = int(time_match[1])
    time_unit = time_match[2].strip()

    return pd.Series({
        "time_num" : time_num,
        "time_unit" : time_unit
        }
    )
```

```python
df['留言時間'].apply(split_time)
```

```python
df[['time_num', 'time_unit']] = df['留言時間'].apply(split_time)
df
```

```python
def get_user_id(text):
    user_id_raw = re.split('/', text)
    user_id = user_id_raw[5]
    return user_id
```

```python
def split_address(raw_address):
    address = re.sub(r"^\d{3,6}", '', raw_address) # 取代開頭郵遞區號3-6個數字成空字串
    match = re.match(r"^(.{2}[縣市])(.+)", address)  
    # (取代開頭2-4任意字元連接縣市，1-3任意字元連接鄉鎮市區為第一群組) (剩下任意字元為第二組)
    if match:
        area = match.group(1)
        detail = match.group(2)
    else:
        area = ''
        detail = address

    return pd.Series({
        "nm_city": area,
        "st_address": detail
        }
    )
```

```python
df['地址'].apply(split_address)
```

```python
df[['nm_city', 'st_address']] = df['地址'].apply(split_address)
df
```

```python
def time_change(row):
    num = row['time_num']
    unit = row['time_unit']
    match unit:
        case '週':
            months_ago = 0.25 * num
            return months_ago
        case '個月':
            months_ago = 1.5 * num
            return months_ago
        case '年':
            months_ago = 18 * num
            return months_ago
        case '天':
            months_ago = 1/30 * num
            return months_ago
        case '小時':
            months_ago = 0
            return months_ago
        case _:
            return None
```

```python
df['months_ago'] = df.apply(time_change, axis=1)
df
```
