```python
import pandas as pd
import re
```

```python
df = pd.read_csv("/workspaces/poetry-demo-test2/clean/longlad_error_edit.csv")
df
```

```python
df["st_url"].tail(5)
```

```python
def get_ld(url):
    
    match = re.search(r"!3d([0-9.]+)!4d([0-9.]+)", url)
    if match:
        longitude = float(match.group(1))
        latitude = float(match.group(2))   
        return pd.Series({
            "longitude" : longitude,
            "latitude" : latitude
            }
        )
    else:
        return pd.Series({
            "longitude" : '',
            "latitude" : ''
            }
        )
```

```python
df["st_url"].apply(get_ld)
```

```python
df[['longitude', 'latitude']] = df["st_url"].apply(get_ld)
df
```

```python
df.to_csv("/workspaces/poetry-demo-test2/clean/store_longlad_error_edit_2.csv", header=True, index=None)
```
